```python
import cv2 as cv
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
model =keras.models.load_model('digits.model')
```

```python
for x in range(1,10):
    img=cv.imread(f'{x}.png')[:,:,0]
    #to invert the color we will use np.invert
    img=np.invert(np.array([img]))
    predict=model.predict(img)
   
    plt.imshow(img[0],cmap=plt.cm.binary)
    plt.show()
    print("The predicted number is",np.argmax(predict))
```
